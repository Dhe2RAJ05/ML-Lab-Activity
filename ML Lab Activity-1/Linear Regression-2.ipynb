```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
ds=pd.read_csv(r"weight.csv")
ds.head()
```

```python
X=ds[["Height","Age"]]
y=ds["Weight"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,              # input and output data
    test_size=0.2,     # 20% for testing
    random_state=42    # extent of randomness
)
```

```python
print("Size of training data: ",len(X_train), "\nSize of test split: ",len(X_test))
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict([[1.8,23]])
y_pred = model.predict(X)
```

```python
plt.scatter(ds['Height'],ds['Weight'], alpha=0.5, color='blue', label='Predicted vs Actual',marker="+")

# Perfect prediction line
plt.plot(ds['Height'],y_pred, color='red', linestyle='--')

# Title and Labels
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
conf_mat=metrics.confusion_matrix(y_test,y_pred)
print("Confusion Matrix: ",conf_mat)
Accuracy_Score=metrics.accuracy_score(y_test,y_pred)
print("Accuracy Score: ",Accuracy_Score)
print("ACcuracy in %: ", Accuracy_Score*100, "%")
```
